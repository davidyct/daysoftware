# Calculate E Value


```python
from math import pi, log

def poisson(e):
    # empirical relationship between modulus of elasticity and poisson's ratio    
    return 0.65 - 0.08 * log(e) / log(10.0)

def ndt3(epav, rad, psi, file_input):
    # Values of subgrade modulus of elasticity
    esubs = [2000.0, 3000.0, 4000.0, 6000.0, 8000.0, 12000.0, 16000.0, 25000.0, 35000.0, 45000.0, 70000.0, 100000.0]

    file_input = open(file_input)
    ndt3_data = file_input.readlines()
    data = []
    for a_line in ndt3_data:
        data.append(eval(a_line))
    #print(data)
    wzh = data[0]
    strh = data[1]
    wz = 2.0 * psi * rad / float(epav)
    
    nesubs = len(esubs)
    ev = -1.0
    if wz > wzh[0][0]:
        ev = esubs[0] * 0.75
    else:
        for i in range(1, nesubs):
            if wz >= wzh[0][i]:
                ev = (esubs[i] - esubs[i - 1]) * (wz - wzh[0][i - 1]) \
                   / (wzh[0][i] - wzh[0][i - 1]) + esubs[i - 1]
                break
    if ev < 0.0:
        ev = esubs[-1] * 1.25
    #print(wz, ev)
    return wz, ev
```

```python
def print_results(sumz, rad): # returns e-value dependent on radius
    epav = 1.0 / (2.0 * rad * sumz * 1e-6)
    #print('SUMZ mils/kip', sumz, 'E-VALUE psi', epav)
    return epav
epav = print_results(0.12654778983261364, 6.0) # Disp1, 6 inch radius
wz, es = ndt3(epav, 6.0, 450.0, 'gels_ndt3_jils_example_results.txt') # falling weight
print('30 ? inch concrete, aggregate base ?, JILS Falling Weight', epav, wz, es)
```
